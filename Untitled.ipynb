```python
def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove stop words
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # pos tag text
    pos_tags = pos_tag(text)
    # lemmatize text
    text = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos(t[1])) for t in pos_tags]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)
```

```python
some['review_clean'] = some['review_body'].apply(lambda x: clean_text(x))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
```

```python
def clean_and_make_scores(filename):
    file_data = pd.read_csv(filename)
    
    print('Cleaning the text')
    some = file_data[['review_id', 'star_rating']]
    some['review_headline_clean'] = file_data['review_headline'].astype(str).apply(lambda x: clean_text(x))
    some['review_body_clean'] = file_data['review_body'].astype(str).apply(lambda x: clean_text(x))
    
    print('Making the scores')
    sid = SentimentIntensityAnalyzer()
    some['sentiment'] = some['review_body_clean'].apply(lambda x: sid.polarity_scores(x))
    some = pd.concat([some.drop(['sentiment'], axis = 1), some['sentiment'].apply(pd.Series)], axis = 1)
    
    return some
```

```python
def safe_list_get (l, idx, default):
    try:
        return l[idx]
    except IndexError:
        return default

def get_adjectives(text):
    text = text.lower()

    
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    
    text = [word for word in text if not any(c.isdigit() for c in word)]
#     stop = stopwords.words('english')
#     text = [x for x in text if x not in stop]
    
    text = [t for t in text if len(t) > 0]
    
    pos_tags = pos_tag(text)
    
    jj_and_rb = set([t for t in pos_tags if t[1] in ['JJ', 'RB'] ])
    
    jj_and_rb = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos(t[1])) for t in jj_and_rb]
    
    return jj_and_rb
```

```python
def get_adjectives_from_file(filename):
    filed = pd.read_csv(filename)
    some = filed[['review_body']]
    
    adjective = pd.DataFrame()
    adjectives  = {}
    for t in some['review_body'].astype(str):
        for i in get_adjectives(t):
            if i not in adjectives:
                adjectives[i] = 1
            else:
                adjectives[i] += 1
                
    adjective['word'] = adjectives.keys()
    adjective['count'] = adjective['word'].apply(lambda x: adjectives[x])
    adjective['compound'] = adjective['word'].apply(lambda x: sid.polarity_scores(x)['compound'])
    
    return adjective
```

```python

def product_id_per_product_parent(filename, plot = True):
    some = pd.read_csv(filename)

    noth = {}
    print(some[some.product_id == pd.unique(some['product_id'])[0]]['product_parent'][0])
    pp =0
    for i in pd.unique(some['product_id']):
    #     print(pp)
        pp = some[some.product_id == i]['product_parent'].iloc[0]
        if pp not in noth:
            noth[pp] = 0
        noth[pp] += 1

    ss = pd.Series(noth)
    if (plot):
        ss.plot(kind='bar')
    return ss
```

```python
microwave_word_list = ['good', 'new', 'perfect', 'easy', 'great', 'never', 'negative', 'bad', 'dead', 'horrible']
hair_dryer_word_list = ['perfect', 'great', 'awesome', 'super', 'powerful', 'bad', 'noisy', 'hard', 'disappointed', 'difficult']
pacifier_word_list = ['awesome', 'great', 'cute', 'clean', 'favorite', 'dirty', 'loose', 'expensive', 'unfortunately', 'wrong']

def make_word_col(filename, word_list):
    some = pd.read_csv(filename)
    some = some[['review_id', 'review_body']]
    
    dic = {}
    for word in word_list:
        dic[word] = []

    for text in some['review_body'].astype(str):
        for word in word_list:
            if word in text:
                dic[word].append(True)
            else:
                dic[word].append(False)
                
    for word in word_list:
        some[word] = pd.Series(dic[word])
        
    return some.drop(['review_body'], axis = 1)
                
    
```

```python
something = make_word_col('cleaned_pacifier.csv', pacifier_word_list)
something.to_csv('pacifier_words.csv')
```
